```python
#Multi indexing
import pandas as pd
#Creating MultiIndex
arrays=[['India','India','USA','USA'],['Delhi','Mumbai','New York','San Francisco']]
index = pd.MultiIndex.from_tuples(list(zip(*arrays)),names=['Country','City'])
#Creating DataFrame
df = pd.DataFrame({'Population':[30,20,15,10]},index=index)
print(df)
```

```python
#Grouping
data={'Department':['HR','HR','IT','IT','Sales','Sales'],
      'Employee':['A','B','C','D','E','F'],
      'Salary':[50000,55000,70000,75000,60000,62000]}
df=pd.DataFrame(data)
#Group by Department and calculate average Salary
grouped_df = df.groupby('Department')['Salary'].mean()
print(grouped_df)
```

```python
#Apply Custom Functions
def salary_category(salary):
    return'High' if salary>60000 else 'Low'
df['Salary Category']=df['Salary'].apply(salary_category)
print(df)
```

```python
#Pivot Tables and CrossTabs
df_pivot = df.pivot_table(index='Department', columns='Salary Category', aggfunc='count')
print(df_pivot)
```

```python
#Pivot Tables and CrossTabs #correct values
df_pivot = df.pivot_table(values='Salary', index='Department', columns='Salary Category', aggfunc='count')
print(df_pivot)
```

```python
#Merging, Joining and Concatenation
#Merging two datasets
df1 = pd.DataFrame({'ID':[1,2,3],'Name':['A','B','C']})
df2 = pd.DataFrame({'ID':[2,3,4],'Salary':[60000,70000,80000]})
df_merged = pd.merge(df1,df2, on='ID', how='outer')
print(df_merged)
```

```python
#Query
df = pd.DataFrame({'A':range(1,1000000),'B':range(2,1000001)})
filtered_df = df.query('A<100')
print(filtered_df.head())
```
